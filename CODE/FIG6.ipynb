```python
import numpy as np
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import loadmat
from datetime import datetime, timedelta
from matplotlib import gridspec

# Step 1: Helper to convert MATLAB datenum to Python datetime
def matlab_datenum_to_datetime(datenum_array):
    datenum_array = np.asarray(datenum_array)
    days = datenum_array % 1
    python_datetimes = [datetime.fromordinal(int(d)) + timedelta(days=float(days[i])) - timedelta(days=366)
                        for i, d in enumerate(datenum_array)]
    return pd.to_datetime(python_datetimes)

def get_season(month):
    if month in [12, 1, 2]:
        return 'DJF'  # Winter
    elif month in [3, 4, 5]:
        return 'MAM'  # Spring
    elif month in [6, 7, 8]:
        return 'JJA'  # Summer
    elif month in [9, 10, 11]:
        return 'SON'  # Autumn
```

```python
# Step 2: Load the MATLAB .mat file
path = '/home/users/anwise/PISCES/DATA/'
files = [path + 'ASPEX02/ASPEX02-3daysfilter-DailyAvg.mat',
         path + 'ASPEX05/ASPEX05-3daysfilter-DailyAvg.mat',
         path + 'ASPEX09/ASPEX09-3daysfilter-DailyAvg.mat']

ds_ls = []

for i_file in files:
    mat = loadmat(i_file)
    
    # Step 3: Extract variables (example assumes variables 'time', 'depth', 'temperature')
    time = mat['tsMoo'].squeeze()  # MATLAB datenum
    depth = mat['depth'].squeeze()  # 1D array
    ual = mat['ualMoof']  # 2D array [time, depth] or similar
    uac = mat['uacMoof']  # 2D array [time, depth] or similar
    
    # Step 4: Convert time to datetime
    time_converted = matlab_datenum_to_datetime(time)
    
    # Step 5: Create xarray Dataset
    ds = xr.Dataset(
        data_vars={
            "ualong": (("time", "depth"), ual),
            "uacross": (("time", "depth"), uac),
            "v": (("time", "depth"), mat['vMoof'].squeeze()),
            "u": (("time", "depth"), mat['uMoof'].squeeze())
        },
        coords={
            "time": time_converted,
            "depth": depth
        },
        attrs={
            "source": "Original " + i_file
        }
    )
        
    months = ds.time.dt.month.values
    seasons = [get_season(m) for m in months]
    
    # Add the season as a coordinate along the time dimension
    ds = ds.assign_coords(season=('time', seasons))

    #ROTATION
    # Stack into 2D array: shape (2, T)
    X = np.vstack([ds.u.sum("depth"), ds.v.sum("depth")])
    
    # Remove mean (detrend) for EOF analysis
    X_demeaned = X - X.mean(axis=1, keepdims=True)
    
    from numpy.linalg import eig
    # Compute 2x2 covariance matrix
    C = np.cov(X_demeaned)
    
    # Eigen-decomposition
    eigvals, eigvecs = eig(C)
    
    # Sort eigenvectors by eigenvalue (descending)
    idx = np.argsort(eigvals)[::-1]
    eigvals = eigvals[idx]
    eigvecs = eigvecs[:, idx]
    
    # First EOF mode
    eof1 = eigvecs[:, 0]
    e_u, e_v = eof1
    
    # Angle in degrees from east (zonal)
    theta_rad = np.arctan2(e_v, e_u)
    theta_deg = np.degrees(theta_rad)
    
    # Wrap angle to [0, 360)
    theta_deg = theta_deg % 360
    
    print(f"Angle of first EOF mode (approx. isobath direction): {theta_deg:.2f}°")
    
    # Rotate original (u, v) into along- and across-isobath components
    ds['u_parallel'] = ds.u * np.cos(theta_rad) + ds.v * np.sin(theta_rad)      # along-isobath
    ds['u_perpendicular'] = -ds.u * np.sin(theta_rad) + ds.v * np.cos(theta_rad)  # across-isobath

    ds_ls.append( ds.u_perpendicular.groupby('season').mean(skipna=True).sel(season=['DJF','MAM','JJA','SON']) )
    #ds_ls.append( -ds.uacross.groupby('season').mean(skipna=True).sel(season=['DJF','MAM','JJA','SON']) )


coord = xr.DataArray( ["ASPEX02", "ASPEX05", "ASPEX09"], dims="mooring", name="mooring" )
da = xr.concat( ds_ls, dim=coord )
ds_ls.append( da.mean("mooring") )
coord = xr.DataArray( ["ASPEX02", "ASPEX05", "ASPEX09", "mooring_mean"], dims="mooring", name="mooring" )
da = xr.concat( ds_ls, dim=coord )
# Save to netcdf
da.to_netcdf( path + "ASPEX02_05_09_mooringmean_u-crossshelf.nc" )
```

```python
def Fig4_plot_season( velocity_obs, velocity_15, velocity_75, v_min, v_max ):
    
    color_ls = ['k', 'tab:blue', 'tab:orange']
    labels_ls = ["a)","b)","c)","d)"]
    
    fig = plt.figure(figsize=(9.8,2.5), dpi=300)
    gs = gridspec.GridSpec( 1, 4, width_ratios=[1,1,1,1], figure=fig, wspace=0.05, hspace=0.05)
    
    for plot_i, season in enumerate( ["DJF","MAM","JJA","SON"] ):
        ax = fig.add_subplot(gs[0,plot_i])
        ax.plot( [0,0], [0,125], "k:" )
        obs = velocity_obs.sel( season=season, mooring="mooring_mean" )
        ax.plot( obs, velocity_obs.depth, 
                    label="mooring \n mean", color=color_ls[0] )
        std_dev = velocity_obs.sel( season=season, mooring=["ASPEX02","ASPEX05","ASPEX09"] ).std("mooring")
        plt.fill_betweenx(velocity_obs.depth, obs - std_dev, obs + std_dev,  color=color_ls[0], alpha=0.1, label='mooring \n ±1 Std Dev',linewidth=0)
        #ax.plot( velocity_obs.sel( season=season, mooring="ASPEX02" ), velocity_obs.depth, 
        #    label="ASPEX02", color=color_ls[1], alpha=0.5, lw=1 )
        #ax.plot( velocity_obs.sel( season=season, mooring="ASPEX05" ), velocity_obs.depth, 
        #    label="ASPEX05", color=color_ls[2], alpha=0.5, lw=1 )
        #ax.plot( velocity_obs.sel( season=season, mooring="ASPEX09" ), velocity_obs.depth, 
        #    label="ASPEX09", color=color_ls[3], alpha=0.5, lw=1 )
        ax.plot( velocity_15.sel( season=season ), velocity_15.depth, 
                    label="AMM15", color=color_ls[1] )
        ax.plot( velocity_75.sel( season=season ), velocity_75.depth, 
                    label="AMM75", color=color_ls[2] )
        ax.set_title(season, fontsize=10, loc='center')
        ax.text(0.02, 0.98, labels_ls[plot_i],
                transform=ax.transAxes,
                fontsize=10, 
                fontweight='bold',
                verticalalignment='top',
                horizontalalignment='left')
        ax.invert_yaxis()
    
        if plot_i != 0:
            ax.set_yticklabels([])
        else:
            ax.set_ylabel('depth [m]', fontsize='medium') 
            #ax.text(-0.35, 0.9, "7.5 km",
            #    transform=ax.transAxes,
            #    fontsize=10, 
            #    fontweight='bold',
            #    verticalalignment='top',
            #    horizontalalignment='left')

        ax.set_xlim(v_min, v_max)
        ax.set_xticks( np.arange(-np.ceil(np.abs(v_min))+0.5, np.ceil(v_max), 0.5) )
        #ax.set_xticklabels([])
        plt.xticks(rotation=45)
        plt.grid(linestyle=':', linewidth=0.2)
        
        # legend
        if plot_i==3:
            ax.legend(bbox_to_anchor=(0.57, 0.65), framealpha=1, fontsize=9)
            ax.set_xlim(v_min, v_max+1)
    
        # ax = fig.add_subplot(gs[1,plot_i])
        # for variable_i, variable in enumerate( velocity_amm15_0["variable"] ):
        #     ax.plot( velocity_amm15_0.sel( season=season, variable=variable )*100, velocity_amm15_0.z_dim, 
        #             label=variable.item(), color=color_ls[variable_i] )
        # ax.text(0.02, 0.98, labels_amm15[plot_i], 
        #     transform=ax.transAxes,
        #     fontsize=10, 
        #     fontweight='bold',
        #     verticalalignment='top',
        #     horizontalalignment='left')
        # ax.invert_yaxis()
        # if plot_i != 0:
        #     ax.set_yticklabels([])
        # else:
        #     ax.set_ylabel('depth [m]', fontsize='medium')
        #     ax.text(-0.35, 0.9, "1.5 km",
        #         transform=ax.transAxes,
        #         fontsize=10, 
        #         fontweight='bold',
        #         verticalalignment='top',
        #         horizontalalignment='left')
        # ax.set_xlim(v_min, v_max) 
        # ax.set_xticks( np.arange(-np.ceil(np.abs(v_min))+1, np.ceil(v_max), 1) )
        # plt.xticks(rotation=45)
        # plt.grid(linestyle=':', linewidth=0.3)
    
        #if plot_i == 2:
        #    ax.set_xlabel('velocity [cm/s]', fontsize='medium')
        #ax.set_title('a) mean (400m)',loc='left')
        plt.rcParams['text.antialiased'] = True
        plt.figtext(0.5, -0.1, 'velocity [cm/s]', ha='center', fontsize='medium', fontweight=300,fontfamily='sans-serif')
    return fig
```
