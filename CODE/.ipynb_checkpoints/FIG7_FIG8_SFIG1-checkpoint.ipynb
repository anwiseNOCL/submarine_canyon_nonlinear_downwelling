```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec

PATH_IN_AMM75 = "/gws/nopw/j04/campus/USERS/anwise/AMM75/AMM75_MOMENTUM/CROSS_SLOPE_FLOW/"
PATH_IN_AMM15 = "/gws/nopw/j04/campus/USERS/anwise/AMM15/AMM15_MOMENTUM/CROSS_SLOPE_FLOW/"
```

```python
amm75_season_ds = xr.open_dataset( PATH_IN_AMM75 + "amm75_mom_seasonal_2015.nc" ).rename({"__xarray_dataarray_variable__":"velocity"})
amm15_season_ds = xr.open_dataset( PATH_IN_AMM15 + "amm15_mom_seasonal_2015.nc" ).rename({"__xarray_dataarray_variable__":"velocity"})
```

```python
def plot_season( velocity_amm75, velocity_amm15, SECTION_idx, v_min, v_max ):
    print( velocity_amm75.section[SECTION_idx].item() )
    velocity_amm75_0 = velocity_amm75.isel(section=SECTION_idx)
    velocity_amm15_0 = velocity_amm15.isel(section=SECTION_idx)
    
    color_ls = ['k', '#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', 
                '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
    labels_amm75 = ["a)","b)","c)","d)"]
    labels_amm15 = ["e)","f)","g)","h)"]
    
    fig = plt.figure(figsize=(9.5,5))
    gs = gridspec.GridSpec( 2, 4, width_ratios=[1,1,1,1], figure=fig, wspace=0.05, hspace=0.05)
    
    for plot_i, season in enumerate( ["DJF","MAM","JJA","SON"] ):
        ax = fig.add_subplot(gs[0,plot_i])
        for variable_i, variable in enumerate( velocity_amm75_0["variable"] ):
            ax.plot( velocity_amm75_0.sel( season=season, variable=variable )*100, velocity_amm75_0.z_dim, 
                    label=variable.item(), color=color_ls[variable_i] )
        ax.set_title(season, fontsize=10, loc='center')
        ax.text(0.02, 0.98, labels_amm75[plot_i],
                transform=ax.transAxes,
                fontsize=10, 
                fontweight='bold',
                verticalalignment='top',
                horizontalalignment='left')
        ax.invert_yaxis()
    
        if plot_i != 0:
            ax.set_yticklabels([])
        else:
            ax.set_ylabel('depth [m]', fontsize='medium') 
            ax.text(-0.35, 0.9, "7.5 km",
                transform=ax.transAxes,
                fontsize=10, 
                fontweight='bold',
                verticalalignment='top',
                horizontalalignment='left')
            
        ax.set_xlim(v_min, v_max)
        ax.set_xticks( np.arange(-np.ceil(np.abs(v_min))+1, np.ceil(v_max), 1) )
        ax.set_xticklabels([])
        plt.grid(linestyle=':', linewidth=0.2)
        
        # legend
        if plot_i==3:
            ax.legend(bbox_to_anchor=(0.8, 1), framealpha=1, fontsize=9)
    
        ax = fig.add_subplot(gs[1,plot_i])
        for variable_i, variable in enumerate( velocity_amm15_0["variable"] ):
            ax.plot( velocity_amm15_0.sel( season=season, variable=variable )*100, velocity_amm15_0.z_dim, 
                    label=variable.item(), color=color_ls[variable_i] )
        ax.text(0.02, 0.98, labels_amm15[plot_i], 
            transform=ax.transAxes,
            fontsize=10, 
            fontweight='bold',
            verticalalignment='top',
            horizontalalignment='left')
        ax.invert_yaxis()
        if plot_i != 0:
            ax.set_yticklabels([])
        else:
            ax.set_ylabel('depth [m]', fontsize='medium')
            ax.text(-0.35, 0.9, "1.5 km",
                transform=ax.transAxes,
                fontsize=10, 
                fontweight='bold',
                verticalalignment='top',
                horizontalalignment='left')
        ax.set_xlim(v_min, v_max) 
        ax.set_xticks( np.arange(-np.ceil(np.abs(v_min))+1, np.ceil(v_max), 1) )
        plt.xticks(rotation=45)
        plt.grid(linestyle=':', linewidth=0.3)
    
        #if plot_i == 2:
        #    ax.set_xlabel('velocity [cm/s]', fontsize='medium')
        #ax.set_title('a) mean (400m)',loc='left')
        fig.text(0.5, 0.02, 'velocity [cm/s]', ha='center', fontsize='medium', fontweight=300,fontfamily='sans-serif'  )#, fontsize='medium')
    return fig
```

```python
velocity_amm75 = amm75_season_ds.velocity.sel(variable=["trd_pvo_e3","trd_hpg_e3","trd_zdf_e3","trd_ldf_e3","trd_keg_e3","trd_rvo_e3","trd_zad_e3"])
velocity_amm75["variable"] = velocity_amm75["variable"].str.slice(start=4, stop=-3)
for term in ['hpg', 'zdf', 'ldf', 'keg', 'rvo', 'zad']:
    velocity_amm75.loc[dict(variable=term)] = -velocity_amm75.sel(variable=term)
velocity_amm75["variable"] = ["total", "pressure gradient", "vertical eddy viscosity", 
                              "lateral eddy viscosity", "lateral advection (non-rot)", "lateral advection (rot)",
                              "vertical advection"]

velocity_amm15 = amm15_season_ds.velocity.sel(variable=["trd_pvo_e3","trd_hpg_e3","trd_zdf_e3","trd_ldf_e3","trd_keg_e3","trd_rvo_e3","trd_zad_e3"])
velocity_amm15["variable"] = velocity_amm15["variable"].str.slice(start=4, stop=-3)
for term in ['hpg', 'zdf', 'ldf', 'keg', 'rvo', 'zad']:
    velocity_amm15.loc[dict(variable=term)] = -velocity_amm15.sel(variable=term)
velocity_amm15["variable"] = ["total", "pressure gradient", "vertical eddy viscosity", 
                              "lateral eddy viscosity", "lateral advection (non-rot)", "lateral advection (rot)",
                              "vertical advection"]
```

```python
# paper Fig 8
fig = plot_season( velocity_amm75, velocity_amm15, 0, -1.75, 1.75 )
#fig.savefig("/home/users/anwise/PISCES/FIGS/" + "cross_200m_velocity_season_celtic_sea.jpg", dpi=300, bbox_inches='tight')
```

```python
def plot_season_aggregated_section( velocity_amm75, velocity_amm15, v_min, v_max ):
    
    color_ls = ['k', '#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', 
                '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
    labels_amm75 = ["a)","b)","c)","d)"]
    labels_amm15 = ["e)","f)","g)","h)"]
    
    fig = plt.figure(figsize=(9.5,5))
    gs = gridspec.GridSpec( 2, 4, width_ratios=[1,1,1,1], figure=fig, wspace=0.05, hspace=0.05)
    
    for plot_i, season in enumerate( ["DJF","MAM","JJA","SON"] ):
        ax = fig.add_subplot(gs[0,plot_i])
        for variable_i, variable in enumerate( velocity_amm75["variable"] ):
            ax.plot( velocity_amm75.sel( season=season, variable=variable )*100, velocity_amm75.z_dim, 
                    label=variable.item(), color=color_ls[variable_i] )
        ax.set_title(season, fontsize=10, loc='center')
        ax.text(0.02, 0.98, labels_amm75[plot_i],
                transform=ax.transAxes,
                fontsize=10, 
                fontweight='bold',
                verticalalignment='top',
                horizontalalignment='left')
        ax.invert_yaxis()
    
        if plot_i != 0:
            ax.set_yticklabels([])
        else:
            ax.set_ylabel('depth [m]', fontsize='medium') 
            ax.text(-0.35, 0.9, "7.5 km",
                transform=ax.transAxes,
                fontsize=10, 
                fontweight='bold',
                verticalalignment='top',
                horizontalalignment='left')
            
        ax.set_xlim(v_min, v_max)
        ax.set_xticks( np.arange(-np.ceil(np.abs(v_min))+1, np.ceil(v_max), 1) )
        ax.set_xticklabels([])
        plt.grid(linestyle=':', linewidth=0.2)
        
        # legend
        if plot_i==3:
            ax.legend(bbox_to_anchor=(0.55, 0.75), framealpha=1, fontsize=9)
    
        ax = fig.add_subplot(gs[1,plot_i])
        for variable_i, variable in enumerate( velocity_amm15["variable"] ):
            ax.plot( velocity_amm15.sel( season=season, variable=variable )*100, velocity_amm15.z_dim, 
                    label=variable.item(), color=color_ls[variable_i] )
        ax.text(0.02, 0.98, labels_amm15[plot_i], 
            transform=ax.transAxes,
            fontsize=10, 
            fontweight='bold',
            verticalalignment='top',
            horizontalalignment='left')
        ax.invert_yaxis()
        if plot_i != 0:
            ax.set_yticklabels([])
        else:
            ax.set_ylabel('depth [m]', fontsize='medium')
            ax.text(-0.35, 0.9, "1.5 km",
                transform=ax.transAxes,
                fontsize=10, 
                fontweight='bold',
                verticalalignment='top',
                horizontalalignment='left')
        ax.set_xlim(v_min, v_max) 
        ax.set_xticks( np.arange(-np.ceil(np.abs(v_min))+1, np.ceil(v_max), 1) )
        plt.xticks(rotation=45)
        plt.grid(linestyle=':', linewidth=0.3)
    
        #if plot_i == 2:
        #    ax.set_xlabel('velocity [cm/s]', fontsize='medium')
        #ax.set_title('a) mean (400m)',loc='left')
        fig.text(0.5, 0.02, 'velocity [cm/s]', ha='center',fontsize='medium', fontweight=300,fontfamily='sans-serif' )#, fontsize='medium')
    return fig
```

```python
# This code will aggregate sections

amm75_sec_lendgth_ds = xr.open_dataset( PATH_IN_AMM75 + "amm75_voltransport_streamfunction_mom_2015.nc" )
amm15_sec_lendgth_ds = xr.open_dataset( PATH_IN_AMM15 + "amm15_voltransport_streamfunction_mom_2015.nc" )

velocity_amm75 = amm75_season_ds.velocity.sel(variable=["trd_pvo_e3","trd_hpg_e3","trd_zdf_e3","trd_ldf_e3","trd_keg_e3","trd_rvo_e3","trd_zad_e3"])
velocity_amm75["variable"] = velocity_amm75["variable"].str.slice(start=4, stop=-3)
for term in ['hpg', 'zdf', 'ldf', 'keg', 'rvo', 'zad']:
    velocity_amm75.loc[dict(variable=term)] = -velocity_amm75.sel(variable=term)
velocity_amm75["variable"] = ["total", "pressure gradient", "vertical eddy viscosity", 
                              "lateral eddy viscosity", "lateral advection (non-rot)", "lateral advection (rot)",
                              "vertical advection"]
section_vel75 = velocity_amm75.sel(section=['West Ireland', 'Rockell', 'Faroe Shetland'])
section_vel75 = section_vel75 * amm75_sec_lendgth_ds.length_section.sel(section=['West Ireland', 'Rockell', 'Faroe Shetland'])
section_vel75 = section_vel75.sum("section") / amm75_sec_lendgth_ds.length_section.sel(section=['West Ireland', 'Rockell', 'Faroe Shetland']).sum("section")

velocity_amm15 = amm15_season_ds.velocity.sel(variable=["trd_pvo_e3","trd_hpg_e3","trd_zdf_e3","trd_ldf_e3","trd_keg_e3","trd_rvo_e3","trd_zad_e3"])
velocity_amm15["variable"] = velocity_amm15["variable"].str.slice(start=4, stop=-3)
for term in ['hpg', 'zdf', 'ldf', 'keg', 'rvo', 'zad']:
    velocity_amm15.loc[dict(variable=term)] = -velocity_amm15.sel(variable=term)
velocity_amm15["variable"] = ["total", "pressure gradient", "vertical eddy viscosity", 
                              "lateral eddy viscosity", "lateral advection (non-rot)", "lateral advection (rot)",
                              "vertical advection"]
section_vel15 = velocity_amm15.sel(section=['West Ireland', 'Rockell', 'Faroe Shetland'])
section_vel15 = section_vel15 * amm15_sec_lendgth_ds.length_section.sel(section=['West Ireland', 'Rockell', 'Faroe Shetland'])
section_vel15 = section_vel15.sum("section") / amm15_sec_lendgth_ds.length_section.sel(section=['West Ireland', 'Rockell', 'Faroe Shetland']).sum("section")
```

```python
# paper Fig 7
fig = plot_season_aggregated_section( section_vel75, section_vel15,-2.3,3.1 )
#fig.savefig("/home/users/anwise/PISCES/FIGS/" + "cross_200m_velocity_season_porc-faro.jpg", dpi=300, bbox_inches='tight')
```
